```python
import matplotlib as plt
import seaborn as sns
import plotly_express as px
```

```python
circuits = pd.read_csv("circuits.csv")
constructor_results = pd.read_csv("constructor_results.csv")
constructor_standings = pd.read_csv("constructor_standings.csv")
constructors = pd.read_csv("constructors.csv")
driver_standings = pd.read_csv("driver_standings.csv")
drivers = pd.read_csv("drivers.csv")
lap_times = pd.read_csv("lap_times.csv")
pi_stops = pd.read_csv("pit_stops.csv")
qualifying = pd.read_csv("qualifying.csv")
races = pd.read_csv("races.csv")
results = pd.read_csv("results.csv")
seasons = pd.read_csv("seasons.csv")
sprint_results = pd.read_csv("sprint_results.csv")
status = pd.read_csv("status.csv")
```

Cell to display the columns in common between the datasets

```python
from collections import defaultdict

datasets = {
    "circuits": circuits,
    "constructor_results": constructor_results,
    "constructor_standings": constructor_standings,
    "constructors": constructors,
    "driver_standings": driver_standings,
    "drivers": drivers,
    "lap_times": lap_times,
    "pit_stops": pi_stops,
    "qualifying": qualifying,
    "races": races,
    "results": results,
    "seasons": seasons,
    "sprint_results": sprint_results,
    "status": status
}

column_map = defaultdict(list)

for name, df in datasets.items():
    for col in df.columns:
        column_map[col].append(name)

column_summary = pd.DataFrame([
    {"Column": col, "Present in datasets": ", ".join(sorted(dsets)), "Count": len(dsets)}
    for col, dsets in sorted(column_map.items())
])

column_summary = column_summary.sort_values(by="Count", ascending=False)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
display(column_summary)
```
